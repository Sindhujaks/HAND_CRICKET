```python
import random

def toss():
    user_choice = input("ODD or EVEN? ").upper()
    user_input = int(input("Enter a number (1-6) for toss: "))
    computer_input = random.randint(1, 6)
    total = user_input + computer_input
    
    if (total % 2 == 0 and user_choice == "EVEN") or (total % 2 == 1 and user_choice == "ODD"):
        user_decision = input("You won the toss! Choose BATTING or BOWLING: ").upper()
        return user_decision
    else:
        return random.choice(["BATTING", "BOWLING"])

def play_hand_cricket():
    status = toss()
    print(f"Your status: {status}")
    score, target, innings = 0, 0, 1
    game_over = False
    
    while not game_over:
        user_input = int(input(f"Your move (1-6): "))
        computer_input = random.randint(1, 6)
        print(f"Computer move: {computer_input}")
        
        if user_input == computer_input:
            if status == "BATTING" and innings == 1:
                print(f"OUT! Your score: {score}. Now you have to bowl.")
                target = score + 1
                status = "BOWLING"
                innings = 2
            elif status == "BOWLING" and innings == 2:
                print(f"OUT! You won! Target was: {target}.")
                game_over = True
            else:
                print("You are out! Computer wins.")
                game_over = True
        else:
            if status == "BATTING":
                score += user_input
                print(f"Your score: {score}")
            else:
                target -= computer_input
                if target <= 0:
                    print("Computer wins!")
                    game_over = True
                else:
                    print(f"Target remaining: {target}")

play_hand_cricket()
```
